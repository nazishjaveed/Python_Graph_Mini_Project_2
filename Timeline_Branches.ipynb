### üåø Create Timeline with Branches using Python

```python
import matplotlib.pyplot as plt

# üìÖ Sample data: timeline events with branching structure
timeline = {
    "main": [(0, 0), (1, 0), (2, 0), (3, 0)],
    "feature1": [(1, 0), (2, 1), (3, 1)],
    "feature2": [(2, 0), (3, -1)]
}

colors = {
    "main": "black",
    "feature1": "blue",
    "feature2": "green"
}

fig, ax = plt.subplots(figsize=(8, 4))

# üñãÔ∏è Draw lines for each branch
for branch, points in timeline.items():
    x, y = zip(*points)
    ax.plot(x, y, marker='o', label=branch, color=colors[branch])

# üõ†Ô∏è Customize plot
ax.set_title("üìÖ Timeline with Branches üåø")
ax.set_xlabel("üïí Time")
ax.set_yticks([-1, 0, 1])
ax.set_yticklabels(["feature2", "main", "feature1"])
ax.set_ylim(-2, 2)
ax.legend(loc='upper left')
ax.grid(True)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
