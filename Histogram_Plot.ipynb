### 📘Create Histogram Plot Using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# 🎲 Generate random data
data = np.random.randn(1000)  # 1000 random values from a normal distribution

# 📊 Create step histogram plot
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, histtype='step', color='blue', linewidth=2)

# 🏷️ Add title and labels
plt.title('📈 Python | Step Histogram Plot')
plt.xlabel('📌 Value')
plt.ylabel('📦 Frequency')
plt.grid(True)

# 👀 Show plot
plt.show()
```
