### üìòCreate Histogram Plot Using Python

```python
import matplotlib.pyplot as plt
import numpy as np

# üé≤ Generate random data
data = np.random.randn(1000)  # 1000 random values from a normal distribution

# üìä Create step histogram plot
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, histtype='step', color='blue', linewidth=2)

# üè∑Ô∏è Add title and labels
plt.title('üìà Python | Step Histogram Plot')
plt.xlabel('üìå Value')
plt.ylabel('üì¶ Frequency')
plt.grid(True)

# üëÄ Show plot
plt.show()
```
