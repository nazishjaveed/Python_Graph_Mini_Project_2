### Create Multivariate Scatter Plot Using Python 🎯📊✨

```python
import numpy as np  # 🧮 For numerical operations
import matplotlib.pyplot as plt  # 📈 For plotting graphs

# 🎲 Random data generation
np.random.seed(0)
x = np.random.rand(100)  # 🔢 Random x-coordinates
y = np.random.rand(100)  # 🔢 Random y-coordinates
colors = np.random.rand(100)  # 🌈 Random colors
sizes = 1000 * np.random.rand(100)  # 🔘 Bubble sizes

# 🎯 Create two different marker types based on a condition
mask = x + y > 1  # ✅ Condition to differentiate markers

plt.figure(figsize=(8, 6))  # 🖼️ Set the figure size

# ⚪ Circle markers for data not satisfying the condition
plt.scatter(x[~mask], y[~mask], s=sizes[~mask], c=colors[~mask], alpha=0.6, cmap='viridis', edgecolors='w')

# 🔺 Triangle markers for data satisfying the condition
plt.scatter(x[mask], y[mask], s=sizes[mask], c=colors[mask], marker='^', alpha=0.6, cmap='viridis', edgecolors='w')

# 📉 Optional: Add a boundary curve for visualization
t = np.linspace(0, 1, 100)
plt.plot(t, 1 - t**2, color='navy')  # 📏 Curve to separate zones

plt.show()  # 👀 Display the final plot
```
