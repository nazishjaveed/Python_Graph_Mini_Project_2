### Create Multivariate Scatter Plot Using Python ðŸŽ¯ðŸ“Šâœ¨

```python
import numpy as np  # ðŸ§® For numerical operations
import matplotlib.pyplot as plt  # ðŸ“ˆ For plotting graphs

# ðŸŽ² Random data generation
np.random.seed(0)
x = np.random.rand(100)  # ðŸ”¢ Random x-coordinates
y = np.random.rand(100)  # ðŸ”¢ Random y-coordinates
colors = np.random.rand(100)  # ðŸŒˆ Random colors
sizes = 1000 * np.random.rand(100)  # ðŸ”˜ Bubble sizes

# ðŸŽ¯ Create two different marker types based on a condition
mask = x + y > 1  # âœ… Condition to differentiate markers

plt.figure(figsize=(8, 6))  # ðŸ–¼ï¸ Set the figure size

# âšª Circle markers for data not satisfying the condition
plt.scatter(x[~mask], y[~mask], s=sizes[~mask], c=colors[~mask], alpha=0.6, cmap='viridis', edgecolors='w')

# ðŸ”º Triangle markers for data satisfying the condition
plt.scatter(x[mask], y[mask], s=sizes[mask], c=colors[mask], marker='^', alpha=0.6, cmap='viridis', edgecolors='w')

# ðŸ“‰ Optional: Add a boundary curve for visualization
t = np.linspace(0, 1, 100)
plt.plot(t, 1 - t**2, color='navy')  # ðŸ“ Curve to separate zones

plt.show()  # ðŸ‘€ Display the final plot
```
