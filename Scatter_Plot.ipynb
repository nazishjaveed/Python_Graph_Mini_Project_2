### Create Scatter Plot Using Python 📌📈

```python
import matplotlib.pyplot as plt  # 🎨 For plotting
import numpy as np  # 🔢 For numerical operations
import pandas as pd  # 🧾 For data handling

# 🎲 Generate random data for population and area
population = np.random.rand(100)  # 👥 Population data (0 to 1)
Area = np.random.randint(100, 600, 100)  # 📏 Area between 100 and 600
continent = ['North America', 'Europe', 'Asia', 'Australia'] * 25  # 🌎 Repeated continent names

# 🧱 Create a DataFrame
df = pd.DataFrame(dict(population=population, Area=Area, continent=continent))  # 📊 Combine into DataFrame

# 🖼️ Create a figure and axis
fig, ax = plt.subplots()  # 🖌️ Initialize plot

# 🎨 Assign colors to continents
colors = {
    'North America': 'red',     # 🌎
    'Europe': 'green',          # 🌍
    'Asia': 'blue',             # 🌏
    'Australia': 'yellow'       # 🦘
}

# 📍 Create scatter plot with mapped continent colors
ax.scatter(df['population'], df['Area'], c=df['continent'].map(colors))  # ✨ Color-coded scatter plot

plt.show()  # 👀 Display the plot
```
