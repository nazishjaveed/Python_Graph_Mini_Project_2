### ðŸ§­ Create Plot of a 2D Vector Field using Python ðŸŒ€ðŸ“Š

```python
# ðŸ“¦ Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# ðŸ§® Creating a grid of x and y values
x, y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))

# ðŸ” Defining vector components (u, v)
u = -y  # âž¡ï¸ Horizontal component
v = x   # â¬†ï¸ Vertical component

# ðŸŽ¯ Creating the vector field plot
plt.figure(figsize=(8, 6))
plt.quiver(x, y, u, v, color='blue')

# ðŸ·ï¸ Adding labels and title
plt.title('ðŸŒªï¸ 2D Vector Field Plot', fontsize=14)
plt.xlabel('ðŸ§­ X-axis', fontsize=12)
plt.ylabel('ðŸ§­ Y-axis', fontsize=12)
plt.grid(True)

# ðŸ‘€ Show the plot
plt.show()
```
