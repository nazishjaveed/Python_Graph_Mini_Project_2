### 🧭 Create Plot of a 2D Vector Field using Python 🌀📊

```python
# 📦 Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 🧮 Creating a grid of x and y values
x, y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))

# 🔁 Defining vector components (u, v)
u = -y  # ➡️ Horizontal component
v = x   # ⬆️ Vertical component

# 🎯 Creating the vector field plot
plt.figure(figsize=(8, 6))
plt.quiver(x, y, u, v, color='blue')

# 🏷️ Adding labels and title
plt.title('🌪️ 2D Vector Field Plot', fontsize=14)
plt.xlabel('🧭 X-axis', fontsize=12)
plt.ylabel('🧭 Y-axis', fontsize=12)
plt.grid(True)

# 👀 Show the plot
plt.show()
```
