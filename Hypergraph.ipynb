### ðŸ“˜ Visualize Hypergraphs with Python

```python
import networkx as nx
import matplotlib.pyplot as plt

# ðŸŒ Example hypergraph structure using dictionary
# Each key is a hyperedge, and the value is a list of nodes connected by it
hyperedges = {
    'e1': ['A', 'B', 'C'],
    'e2': ['B', 'D'],
    'e3': ['C', 'D', 'E'],
    'e4': ['A', 'E']
}

# ðŸ”— Create a bipartite graph: nodes (normal) + hyperedges (as special nodes)
B = nx.Graph()

# âž• Add nodes and hyperedges as separate types
for edge, nodes in hyperedges.items():
    B.add_node(edge, type='hyperedge')  # hyperedge node
    for node in nodes:
        B.add_node(node, type='node')
        B.add_edge(edge, node)

# ðŸ—ºï¸ Positioning: use bipartite layout
pos = nx.spring_layout(B, seed=42)

# ðŸŽ¨ Color map: distinguish between node types
colors = ['red' if B.nodes[n].get('type') == 'hyperedge' else 'skyblue' for n in B.nodes]

# ðŸŽ¥ Draw the bipartite graph
plt.figure(figsize=(8, 6))
nx.draw(B, pos, with_labels=True, node_color=colors, node_size=800, font_weight='bold')
plt.title("ðŸŒŸ Hypergraph Visualization in Python ðŸŒŸ")
plt.show()
```
