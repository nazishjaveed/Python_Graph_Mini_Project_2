###   üì¶Create Network Grapgh Using Python

```python
import networkx as nx  # üì¶ For network graph handling
import matplotlib.pyplot as plt  # üìä For plotting
import numpy as np  # üî¢ For numerical computations

# üï∏Ô∏è Create a graph
G = nx.Graph()

# üìå Add nodes
nodes = ['a', 'b', 'c', 'd', 'e', 'f']
G.add_nodes_from(nodes)

# üåà Add weighted edges (edge, weight)
edges = [
    ('a', 'b', 4), ('a', 'c', 3), ('a', 'd', 2),
    ('b', 'c', 5), ('b', 'e', 2), ('b', 'f', 1),
    ('c', 'd', 4), ('c', 'e', 3), ('c', 'f', 2),
    ('d', 'f', 4)
]
G.add_weighted_edges_from(edges)

# üé® Create a color map based on weights
weights = [data['weight'] for u, v, data in G.edges(data=True)]
colors = [plt.cm.viridis(weight / max(weights)) for weight in weights]  # Normalized colormap

# üìç Position nodes in circular layout
pos = nx.spring_layout(G, seed=42)

# üñºÔ∏è Draw nodes
nx.draw_networkx_nodes(G, pos, node_color='deepskyblue', node_size=800)

# üñçÔ∏è Draw edges with color variation
nx.draw_networkx_edges(G, pos, edge_color=colors, width=2)

# üè∑Ô∏è Draw labels
nx.draw_networkx_labels(G, pos, font_size=12, font_color='white')

# üåë Dark background
plt.gca().set_facecolor('black')
plt.axis('off')  # üö´ Hide axis
plt.show()
```
