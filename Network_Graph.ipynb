###   📦Create Network Grapgh Using Python

```python
import networkx as nx  # 📦 For network graph handling
import matplotlib.pyplot as plt  # 📊 For plotting
import numpy as np  # 🔢 For numerical computations

# 🕸️ Create a graph
G = nx.Graph()

# 📌 Add nodes
nodes = ['a', 'b', 'c', 'd', 'e', 'f']
G.add_nodes_from(nodes)

# 🌈 Add weighted edges (edge, weight)
edges = [
    ('a', 'b', 4), ('a', 'c', 3), ('a', 'd', 2),
    ('b', 'c', 5), ('b', 'e', 2), ('b', 'f', 1),
    ('c', 'd', 4), ('c', 'e', 3), ('c', 'f', 2),
    ('d', 'f', 4)
]
G.add_weighted_edges_from(edges)

# 🎨 Create a color map based on weights
weights = [data['weight'] for u, v, data in G.edges(data=True)]
colors = [plt.cm.viridis(weight / max(weights)) for weight in weights]  # Normalized colormap

# 📍 Position nodes in circular layout
pos = nx.spring_layout(G, seed=42)

# 🖼️ Draw nodes
nx.draw_networkx_nodes(G, pos, node_color='deepskyblue', node_size=800)

# 🖍️ Draw edges with color variation
nx.draw_networkx_edges(G, pos, edge_color=colors, width=2)

# 🏷️ Draw labels
nx.draw_networkx_labels(G, pos, font_size=12, font_color='white')

# 🌑 Dark background
plt.gca().set_facecolor('black')
plt.axis('off')  # 🚫 Hide axis
plt.show()
```
