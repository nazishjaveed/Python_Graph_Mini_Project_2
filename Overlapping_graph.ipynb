### üìò Create Overlapping Graphs Using Python

```python
import matplotlib.pyplot as plt

# üìä Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]      # üìà Line 1
y2 = [1, 3, 5, 7, 9]       # üìâ Line 2
y3 = [5, 4, 3, 2, 1]       # ‚ûñ Line 3

# Plotting multiple lines on the same graph
plt.plot(x, y1, label='Line 1', color='blue', marker='o')  # üîµ Line 1 (Circle markers)
plt.plot(x, y2, label='Line 2', color='green', marker='s')  # üü© Line 2 (Square markers)
plt.plot(x, y3, label='Line 3', color='red', linestyle='--', marker='^')  # üî¥ Line 3 (Triangle markers)

# Adding title and labels
plt.title('Overlapping Graphs using Matplotlib üìà')
plt.xlabel('X Axis üìä')
plt.ylabel('Y Axis üìè')

# Show legend and grid
plt.legend()
plt.grid(True)

# Show the graph
plt.show()
```
