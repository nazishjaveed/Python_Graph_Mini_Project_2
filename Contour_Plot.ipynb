### ðŸ“˜Create Contour Plot Using Python

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Creating sample data for the contour plot
x = np.linspace(-5.0, 5.0, 100)
y = np.linspace(-5.0, 5.0, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # Z is a function of X and Y

# Plotting the contour plot
plt.figure(figsize=(8, 6))
cp = plt.contour(X, Y, Z, 20, cmap='viridis')  # 20 contour levels and 'viridis' colormap

# Adding color bar
plt.colorbar(cp)

# Adding title and labels
plt.title('Contour Plot in Python', fontsize=14)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

# Show plot
plt.show()
```
