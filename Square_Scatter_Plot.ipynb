### üü©üìä Create Square Scatter Plot Using Python üêçüé®

```python
import matplotlib.pyplot as plt
import numpy as np

# üå±üìä Generate random data
np.random.seed(0)
x = np.random.rand(60) * 50
y = np.random.rand(60) * 50
colors = np.random.rand(60) * 50  # üé® Color scale
sizes = np.random.rand(60) * 1000  # üî≤ Size scale for squares

# üñºÔ∏è Create the plot
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x, y, s=sizes, c=colors, cmap='summer', marker='s', alpha=0.8)

# üå°Ô∏è Add colorbar
plt.colorbar(scatter)

# üè∑Ô∏è Add title
plt.title("üü© Square Scatter Plot Example")

# üëÄ Show plot
plt.show()
```
