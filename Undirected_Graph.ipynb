### Create Undirected Connected Graph ✅ using python

```python
import networkx as nx  # 📦 For graph creation and manipulation
import matplotlib.pyplot as plt  # 📊 For plotting

# 🧱 Create a graph object
G = nx.Graph()

# 🔘 Add nodes
nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
G.add_nodes_from(nodes)

# 🔗 Add edges based on the image
edges = [
    ('A', 'D'), ('A', 'G'), ('A', 'B'),
    ('B', 'G'), ('B', 'C'),
    ('G', 'C'), ('G', 'F'),
    ('C', 'F'), ('C', 'E'),
    ('D', 'F'), ('F', 'E')
]
G.add_edges_from(edges)

# 🎨 Draw the graph
pos = nx.spring_layout(G, seed=42)  # 🧲 Consistent layout positioning
nx.draw(
    G, pos, with_labels=True, 
    node_color='lightgreen',  # 🌿 Node color
    node_size=1000,  # 📏 Node size
    font_weight='bold', 
    edge_color='gray'  # 🌫️ Edge color
)

plt.title("📈 Graph Representation 🕸️")
plt.show()  # 👀 Display the plot
```
