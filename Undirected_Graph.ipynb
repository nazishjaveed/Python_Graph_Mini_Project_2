### Create Undirected Connected Graph ‚úÖ using python

```python
import networkx as nx  # üì¶ For graph creation and manipulation
import matplotlib.pyplot as plt  # üìä For plotting

# üß± Create a graph object
G = nx.Graph()

# üîò Add nodes
nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
G.add_nodes_from(nodes)

# üîó Add edges based on the image
edges = [
    ('A', 'D'), ('A', 'G'), ('A', 'B'),
    ('B', 'G'), ('B', 'C'),
    ('G', 'C'), ('G', 'F'),
    ('C', 'F'), ('C', 'E'),
    ('D', 'F'), ('F', 'E')
]
G.add_edges_from(edges)

# üé® Draw the graph
pos = nx.spring_layout(G, seed=42)  # üß≤ Consistent layout positioning
nx.draw(
    G, pos, with_labels=True, 
    node_color='lightgreen',  # üåø Node color
    node_size=1000,  # üìè Node size
    font_weight='bold', 
    edge_color='gray'  # üå´Ô∏è Edge color
)

plt.title("üìà Graph Representation üï∏Ô∏è")
plt.show()  # üëÄ Display the plot
```
