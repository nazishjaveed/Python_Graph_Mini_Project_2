### Create sinusoidal plots using Python üêçüìà

```python
import matplotlib.pyplot as plt
import numpy as np

# üìä Generate X values
x = np.linspace(0, 10, 25)

# üåä Generate Y values
y1 = np.sin(x)  # üîµ Sine wave
y2 = np.cos(x)  # üî¥ Cosine wave

# üé® Create the plot
plt.plot(x, y1, 'bo-', label='Line 1 (sin) üîµ')   # blue circle markers with solid line
plt.plot(x, y2, 'r*:', label='Line 2 (cos) üî¥')   # red star markers with dotted line

# üè∑Ô∏è Add labels and title
plt.xlabel('X-axis ‚û°Ô∏è')
plt.ylabel('Y-axis ‚¨ÜÔ∏è')
plt.title('Sinusoidal Waves üåäüìà')

# üìå Add legend
plt.legend()

# üìê Show grid and plot
plt.grid(True)
plt.show()
```
