### üåêüìÖ Create 3D Surface Plot: Wavelength üåà vs. Date üìÜ vs. Intensity üî•

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.dates as mdates
import datetime

# Generate sample wavelength and date arrays
wavelengths = np.linspace(6520, 6580, 100)  # Wavelengths in √Öngstr√∂m
dates = [datetime.datetime(2020, 12, 1) + datetime.timedelta(days=i) for i in range(40)]
date_nums = mdates.date2num(dates)

# Create 2D grid
X, Y = np.meshgrid(wavelengths, date_nums)

# Simulated intensity values with some peaks
Z = np.exp(-0.01 * (X - 6550)**2) * np.exp(-0.005 * (Y - mdates.date2num(datetime.datetime(2020, 12, 20)))**2)
Z += 0.1 * np.random.rand(*X.shape)  # Add noise

# Plotting
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Format date axis
ax.yaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
ax.set_ylabel('Date')
ax.set_xlabel('Wavelength (√Ö)')
ax.set_zlabel('Intensity')

# Add color bar and title
fig.colorbar(surf, shrink=0.5, aspect=10)
ax.set_title('3D Surface Plot: Wavelength vs. Date vs. Intensity')

plt.tight_layout()
plt.show()
```
