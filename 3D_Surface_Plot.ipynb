### 🔢 Create 3D Surface Plot Using Python

```python
import numpy as np  # 🔢 For numerical operations
import matplotlib.pyplot as plt  # 📊 For plotting
from mpl_toolkits.mplot3d import Axes3D  # 🧭 For 3D plots

# 🌐 Create grid data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# 🔺 Define a smooth function and add "bumps" using random noise
z = np.exp(-0.1 * (x**2 + y**2))  # Gaussian bump 🌄
z += 0.05 * np.random.rand(*x.shape)  # Add random noise 🎲

# 🎨 Plotting the surface
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 🌈 Create surface plot with colormap
surf = ax.plot_surface(x, y, z, cmap='jet', edgecolor='none')

# 🎚️ Add color bar and axis labels
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
ax.set_title('3D Surface Plot with Bumps 🌋')
ax.set_xlabel('X ➡️')
ax.set_ylabel('Y ⬆️')
ax.set_zlabel('Z ↕️')

# 📷 Show the plot
plt.show()
```
