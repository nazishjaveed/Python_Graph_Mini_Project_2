### ü•ã Create Zachary's Karate Club Graph Using Python üêçüìä

```python
import networkx as nx  # üì¶ Network analysis library
import matplotlib.pyplot as plt  # üìä For plotting graphs
from networkx.algorithms import community  # üë• Community detection tools
import numpy as np  # üî¢ Numerical operations

# ü•ã Load the Karate Club graph
G = nx.karate_club_graph()

# üß† Detect communities using the greedy modularity algorithm
communities = community.greedy_modularity_communities(G)

# üó∫Ô∏è Create a mapping from node to community index
node_color_map = {}
for i, com in enumerate(communities):
    for node in com:
        node_color_map[node] = i

# üé® Assign colors based on community
colors = [node_color_map[node] for node in G.nodes()]

# üñºÔ∏è Draw the graph
plt.figure(figsize=(8, 5))
pos = nx.spring_layout(G, seed=42)  # üß≤ Positioning nodes for better layout
nx.draw(G, pos, with_labels=True, node_color=colors, cmap=plt.cm.Set1, node_size=500)
plt.title("ü•ã Zachary's Karate Club - Community Detection üë•")
plt.show()  # üëÄ Display the plot
```
