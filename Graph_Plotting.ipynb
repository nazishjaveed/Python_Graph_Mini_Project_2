### üìòCreate Line Graph Using Pthon

```python
import matplotlib.pyplot as plt

# üìä Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 7]

# üìà Create the plot
plt.plot(x, y, marker='o', color='blue', linestyle='--', label='Data Line')

# üè∑Ô∏è Add title and labels
plt.title('Simple Line Graph üìâ')
plt.xlabel('X-axis üìê')
plt.ylabel('Y-axis üìè')

# üé® Show legend
plt.legend()

# üß≠ Display the plot
plt.grid(True)
plt.show()
```
