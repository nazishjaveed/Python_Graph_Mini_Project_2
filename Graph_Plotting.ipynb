### 📘Create Line Graph Using Pthon

```python
import matplotlib.pyplot as plt

# 📊 Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 7]

# 📈 Create the plot
plt.plot(x, y, marker='o', color='blue', linestyle='--', label='Data Line')

# 🏷️ Add title and labels
plt.title('Simple Line Graph 📉')
plt.xlabel('X-axis 📐')
plt.ylabel('Y-axis 📏')

# 🎨 Show legend
plt.legend()

# 🧭 Display the plot
plt.grid(True)
plt.show()
```
