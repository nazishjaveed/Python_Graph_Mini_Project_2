### 📘Create  2D-Plotting Using Python

```python
# 2D-Plotting | Data Science with Python 📊

import matplotlib.pyplot as plt
import numpy as np

# Generating data for 2D plot 📉
x = np.linspace(0, 10, 100)           # 100 values from 0 to 10 🔢
y = np.sin(x)                         # y = sin(x) 🔄

# Create the 2D line plot 📈
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='purple', linewidth=2, label='sin(x)')

# Adding title and labels 🏷️
plt.title('2D-Plotting | Data Science with Python 📊', fontsize=14)
plt.xlabel('X Axis (radians) 🔢', fontsize=12)
plt.ylabel('Y Axis (amplitude) 🎵', fontsize=12)

# Grid, legend, and display 🔲
plt.grid(True)
plt.legend()
plt.show()
```
