### ðŸ“˜Create  2D-Plotting Using Python

```python
# 2D-Plotting | Data Science with Python ðŸ“Š

import matplotlib.pyplot as plt
import numpy as np

# Generating data for 2D plot ðŸ“‰
x = np.linspace(0, 10, 100)           # 100 values from 0 to 10 ðŸ”¢
y = np.sin(x)                         # y = sin(x) ðŸ”„

# Create the 2D line plot ðŸ“ˆ
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='purple', linewidth=2, label='sin(x)')

# Adding title and labels ðŸ·ï¸
plt.title('2D-Plotting | Data Science with Python ðŸ“Š', fontsize=14)
plt.xlabel('X Axis (radians) ðŸ”¢', fontsize=12)
plt.ylabel('Y Axis (amplitude) ðŸŽµ', fontsize=12)

# Grid, legend, and display ðŸ”²
plt.grid(True)
plt.legend()
plt.show()
```
