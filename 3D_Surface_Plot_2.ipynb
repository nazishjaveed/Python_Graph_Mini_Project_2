### ðŸŽ¯ Create 3D Scatter Plot With Rainbow ColorMap ðŸŒˆðŸ“ŠðŸ§­

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ðŸŽ² Generate random data
np.random.seed(0)
x = np.random.standard_normal(100) * 10
y = np.random.standard_normal(100) * 10
z = np.random.standard_normal(100) * 100

# ðŸŒˆ Color mapping based on Z values
colors = z

# ðŸ“Š Create a 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=colors, cmap='rainbow')

# ðŸŽ¨ Add colorbar
cb = fig.colorbar(scatter, ax=ax, shrink=0.6)
cb.set_label('Z-axis ðŸ”¼')

# ðŸ·ï¸ Set labels
ax.set_title('ðŸŒˆ 3D Scatter Plot in Python ðŸ“')
ax.set_xlabel('X-axis âž¡ï¸')
ax.set_ylabel('Y-axis â¬†ï¸')
ax.set_zlabel('Z-axis ðŸ”¼')

plt.tight_layout()
plt.show()
```
