### üìò Create Graph Plotting Using Python

```python
# Importing required library üìö
import matplotlib.pyplot as plt

# Sample data üìä
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

# Creating the figure and plotting the graph üìà
plt.figure(figsize=(8, 5))  # Set the figure size
plt.plot(x, y, color='blue', marker='o', linestyle='-', linewidth=2, markersize=8)

# Adding titles and labels üè∑Ô∏è
plt.title('Graph Plotting in Python | Board Infinity üìä', fontsize=14)
plt.xlabel('X-Axis üìê', fontsize=12)
plt.ylabel('Y-Axis üìè', fontsize=12)

# Enabling grid üî≤
plt.grid(True)

# Display the graph üëÄ
plt.show()
```
