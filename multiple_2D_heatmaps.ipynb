### üî• Create Multiple 2D Heatmaps Using Python üêçüìäüó∫Ô∏è

```python
import numpy as np  # üî¢ Numerical computations
import matplotlib.pyplot as plt  # üìä Plotting
from mpl_toolkits.mplot3d import Axes3D  # üß± 3D plotting tools

# üìê Create a grid of x, y values
x = y = np.linspace(0, 1, 100)
x, y = np.meshgrid(x, y)

# üåä Define three functions to plot on different z levels
z1 = np.sin(4 * np.pi * x) * np.sin(4 * np.pi * y)  # üîù Top layer
z2 = x * y                                          # ‚öñÔ∏è Middle layer
z3 = x + y                                          # üîΩ Bottom layer

# üìè Set z-levels for each surface
z_levels = [1.0, 0.6, 0.2]

# üé® Plotting
fig = plt.figure(figsize=(6, 10))
ax = fig.add_subplot(111, projection='3d')  # üì¶ 3D plot setup

# üñºÔ∏è Plot each surface with color mapping
ax.plot_surface(
    x, y, z_levels[0] + 0*z1,
    facecolors=plt.cm.jet((z1 - z1.min()) / (z1.max() - z1.min())),
    rstride=1, cstride=1, antialiased=False
)
ax.plot_surface(
    x, y, z_levels[1] + 0*z2,
    facecolors=plt.cm.jet((z2 - z2.min()) / (z2.max() - z2.min())),
    rstride=1, cstride=1, antialiased=False
)
ax.plot_surface(
    x, y, z_levels[2] + 0*z3,
    facecolors=plt.cm.jet((z3 - z3.min()) / (z3.max() - z3.min())),
    rstride=1, cstride=1, antialiased=False
)

# üßΩ Clean up the plot appearance
ax.set_xticks([0, 0.5, 1.0])  # üìç X-ticks
ax.set_yticks([0, 0.5, 1.0])  # üìç Y-ticks
ax.set_zticks(z_levels)       # üìç Z-ticks
ax.view_init(elev=30, azim=-60)  # üîÑ View angle

plt.tight_layout()
plt.show()  # üëÄ Show the final 3D plot
```
