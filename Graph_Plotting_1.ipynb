### 📘 Create Graph Plotting Using Python

```python
# Importing required library 📚
import matplotlib.pyplot as plt

# Sample data 📊
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

# Creating the figure and plotting the graph 📈
plt.figure(figsize=(8, 5))  # Set the figure size
plt.plot(x, y, color='blue', marker='o', linestyle='-', linewidth=2, markersize=8)

# Adding titles and labels 🏷️
plt.title('Graph Plotting in Python | Board Infinity 📊', fontsize=14)
plt.xlabel('X-Axis 📐', fontsize=12)
plt.ylabel('Y-Axis 📏', fontsize=12)

# Enabling grid 🔲
plt.grid(True)

# Display the graph 👀
plt.show()
```
