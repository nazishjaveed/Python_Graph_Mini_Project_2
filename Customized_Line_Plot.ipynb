### Create Customized Line Plot Using Python

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 1, 5, 2, 6]

# Create the plot
plt.plot(x, y, 'o--', color='green', markerfacecolor='blue', markersize=10)

# Add labels and title
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Some cool customizations!')

# Show the plot
plt.grid(True)
plt.show()
```
