```python
# Author: Dmytro
# Topic: Graphs with Python ðŸ“Š

import matplotlib.pyplot as plt

# Sample data ðŸ“ˆ
x = [1, 2, 3, 4, 5]
y_line = [1, 4, 9, 16, 25]  # ðŸ“ˆ Line Graph Data
y_bar = [5, 7, 3, 8, 6]      # ðŸ“Š Bar Graph Data

# Plot 1: Line Graph ðŸ“‰
plt.figure(figsize=(10, 5))
plt.plot(x, y_line, label='Line Graph', color='blue', marker='o')
plt.title('Line Graph Example ðŸ“ˆ')
plt.xlabel('X Axis ðŸ“')
plt.ylabel('Y Axis ðŸ“')
plt.legend()
plt.grid(True)
plt.show()

# Plot 2: Bar Graph ðŸ“Š
plt.figure(figsize=(10, 5))
plt.bar(x, y_bar, color='orange', label='Bar Graph')
plt.title('Bar Graph Example ðŸ“Š')
plt.xlabel('X Axis ðŸ“')
plt.ylabel('Values ðŸ“')
plt.legend()
plt.grid(True, axis='y')
plt.show()

# Plot 3: Scatter Plot ðŸ”µ
plt.figure(figsize=(10, 5))
plt.scatter(x, y_line, color='green', label='Scatter Plot')
plt.title('Scatter Plot Example ðŸ”´')
plt.xlabel('X Axis ðŸ“')
plt.ylabel('Y Axis ðŸ“')
plt.legend()
plt.grid(True)
plt.show()
```
