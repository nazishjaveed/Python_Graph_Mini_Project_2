### 📘Create Line Graph Graphs Using Python | by Dmytro

```python
# Author: Dmytro
# Topic: Graphs with Python 📊

import matplotlib.pyplot as plt

# Sample data 📈
x = [1, 2, 3, 4, 5]
y_line = [1, 4, 9, 16, 25]  # 📈 Line Graph Data
y_bar = [5, 7, 3, 8, 6]      # 📊 Bar Graph Data

# Plot 1: Line Graph 📉
plt.figure(figsize=(10, 5))
plt.plot(x, y_line, label='Line Graph', color='blue', marker='o')
plt.title('Line Graph Example 📈')
plt.xlabel('X Axis 📐')
plt.ylabel('Y Axis 📏')
plt.legend()
plt.grid(True)
plt.show()

# Plot 2: Bar Graph 📊
plt.figure(figsize=(10, 5))
plt.bar(x, y_bar, color='orange', label='Bar Graph')
plt.title('Bar Graph Example 📊')
plt.xlabel('X Axis 📐')
plt.ylabel('Values 📍')
plt.legend()
plt.grid(True, axis='y')
plt.show()

# Plot 3: Scatter Plot 🔵
plt.figure(figsize=(10, 5))
plt.scatter(x, y_line, color='green', label='Scatter Plot')
plt.title('Scatter Plot Example 🔴')
plt.xlabel('X Axis 📐')
plt.ylabel('Y Axis 📏')
plt.legend()
plt.grid(True)
plt.show()
```
