### üåà Create Rainbow Graph Using Python üêçüìà

```python
import matplotlib.pyplot as plt  # üìä
import numpy as np  # üî¢

x = np.linspace(0, 100, 50)  # ‚û∞ Generate 50 points between 0 and 100

y_old = -(x - 50) ** 2 + 2500  # üü¢ Initial parabola
for delta, color in zip([2250, 2000, 1750, 1500, 1250, 1000], ["r", "orange", "g", "b", "indigo", "violet"]):
    y_new = -(x - 50) ** 2 + delta  # ‚ûï Shift parabola upward
    plt.plot(x, y_old, "-k")  # ‚ö´ Draw outline of each band
    plt.fill_between(x, y_old, y_new, color=color)  # üåà Fill between parabolas with color
    y_old = y_new  # üîÅ Update for next band

plt.ylim(0, 2500)  # üìè Set y-axis limit
plt.show()  # üëÄ Display plot
```
